```python
# date 07/04/2026
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
def alice_unknown_state_prep(tpc):
    tpc.h(0)
    tpc.z(0) 
    tpc.h(0)

def bell_state_prep(tpc):  
    tpc.h(1)
    tpc.cx(1,2)            

def alice_measuring(tpc):
    tpc.measure(0,0)
    tpc.measure(1,1)

    aer = AerSimulator()
    counts = aer.run(tpc,shots = 1).result().get_counts()
    msmts = list(counts.keys())[0] 
    return msmts

def bob_treating(tpc,msmts):
    bit1,bit2 = int(msmts[0]),int(msmts[1]) 
    if bit1 and bit2:
        tpc.z(2)
        tpc.x(2)
    elif bit1:
        tpc.z(2)
    elif bit2:
        tpc.x(2) 
    
    
# def bob_ 
tpc = QuantumCircuit(3,2)

alice_unknown_state_prep(tpc) 
print("after alice state prep")
display(tpc.draw("mpl")) 

bell_state_prep(tpc)
print("after bell state")
display(tpc.draw("mpl"))

msmts = alice_measuring(tpc)   
print("after alice measuring")
print("measurement outcomes: ",msmts)
display(tpc.draw("mpl"))

bob_treating(tpc,msmts)
print("after bob treating")
display(tpc.draw("mpl"))
```
