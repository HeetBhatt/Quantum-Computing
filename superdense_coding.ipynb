```python
from qiskit import QuantumCircuit
```

```python
def alice_data_encoding(sdc,bit1,bit2):
    sdc.h(0)
    sdc.cx(0,1)
    if bit1 and bit2:
        sdc.z(0)
        sdc.x(0)
    elif bit1:
        sdc.z(0)
    elif bit2:
        sdc.x(0) 

def bob_data_decoding(sdc): 
    sdc.cx(0,1)
    sdc.h(0)
    sdc.measure(0,0)
    sdc.measure(1,1)
 
bits = input("Enter 2 bits to send from alice to bob\n")

if len(bits) != 2 or any(b not in '01' for b in bits):
    print("Invalid input. Enter exactly 2 bits (0 or 1).")
else:
    # Individual bits
    bit1 = int(bits[0])
    bit2 = int(bits[1])

    print("First bit:", bit1)
    print("Second bit:", bit2)
sdc = QuantumCircuit(2,2)
alice_data_encoding(sdc,bit1,bit2)  
bob_data_decoding(sdc)
sdc.draw("mpl") 
```
